```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

#############################################################
#Création d'un nouveau dataset
#############################################################
def creation_data():
  df=pd.DataFrame(columns=['x','y','couleur'])
  pixel_x=640
  pixel_y=480
  taille_dataset=800
  x=np.random.randint(pixel_x, size=taille_dataset)
  y=np.random.randint(pixel_y, size=taille_dataset)
  df['x']=x
  df['y']=y


  for i in range(len(df)):
      if(df['x'][i]in range(160,480) and df['y'][i]in range(120,360)):
        df['couleur'][i]='red'
      else:
         df['couleur'][i]='blue'

  return df


```

```python
dataset=creation_data()
dataset
```

```python
dataset[dataset['couleur']=='red']
```

```python
plt.scatter(dataset['x'],dataset['y'],color=dataset['couleur'])
```

```python

XY = dataset.values[:,:2]
print('XY = \n', XY,'\n')

C = dataset.values[:,2]
print('C = \n', C,'\n')

from sklearn import model_selection
XY_Train,XY_Test,C_Train,C_Test = model_selection.train_test_split(XY,C,test_size=0.1, random_state=20)
print('XY_Train = \n', XY_Train,'\n')
print('XY_Test = \n', XY_Test,'\n')
print('C_Train = \n', C_Train,'\n')
print('C_Test = \n', C_Test,'\n')
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

scaler.fit(XY_Train)
print('Data max de XY_Train :\n', scaler.data_max_, '\n')
XY_Train = scaler.transform(XY_Train)
print('Normalisation de XY_Train :\n', XY_Train, '\n')

scaler.fit(XY_Test)
print('Data max de XY_Test :\n', scaler.data_max_, '\n')
XY_Test = scaler.transform(XY_Test)
print('Normalisation de XY_Test :\n', XY_Test, '\n')
```

```python

from sklearn.neural_network import MLPClassifier
NN = MLPClassifier(hidden_layer_sizes=(2),activation="logistic",solver="lbfgs")
Fractional_shift_number = 5
Biais_shift_number = 8
#apprentissage
##############################################################
NN.fit(XY_Train,C_Train)

##############################################################
#affichage des coefficients
##############################################################
print('Poids synaptiques = \n', NN.coefs_,'\n')
print('Biais = \n', NN.intercepts_,'\n')

##############################################################
#vérification de l'apprentissage
##############################################################
pred = NN.predict(XY_Test) #prédiction sur l'échantillon test
print('Prediction = \n', pred, '\n')
from sklearn import metrics
print(metrics.confusion_matrix(C_Test,pred)) #mesure des performances
print("Taux d'erreur = " + str(1-metrics.accuracy_score(C_Test,pred)), '\n')

##############################################################
#Conversion des coefficients de la couche de sortie en virgule fixe
##############################################################
# Les poids synaptiques et le biais du neurone de la couche de sortie sont exprimés en nombre à virgule flottante
# Le FPGA dans lequel chaque poids et biais va être implanté travaille avec des nombres entiers.
# On pourrait se contenter de n'utiliser que la partie entière de chaque poids et biais ... au détriment de la précision
# Mais on peut également choisir d'utiliser la partie fractionnaire
# Pour ce faire, on multiplie chaque poids synaptique et biais par un nombre puissance de 2 (ici 2^5)
# Cela augmentera la précision du neurone
# Mais cela augmentera également la taille binaire de chaque poids et biais à stocker en mémoire


# définition des neurones


Output_Layer_Coefs = [[0,0,0,0],[0,0]]
Output_Layer_Biais = [[0,0],[0]]
print(NN.intercepts_[1][0])
Output_Layer_Coefs[0][0] = int((NN.coefs_[0][0][0]) * 2**Fractional_shift_number)
Output_Layer_Coefs[0][1] = int((NN.coefs_[0][0][1]) * 2**Fractional_shift_number)
Output_Layer_Coefs[0][2] = int((NN.coefs_[0][1][0]) * 2**Fractional_shift_number)
Output_Layer_Coefs[0][3] = int((NN.coefs_[0][1][1]) * 2**Fractional_shift_number)

Output_Layer_Coefs[1][0] = int((NN.coefs_[1][0][0]) * 2**Fractional_shift_number)
Output_Layer_Coefs[1][1] = int((NN.coefs_[1][1][0]) * 2**Fractional_shift_number)

Output_Layer_Biais[0][0] = int((NN.intercepts_[0][0]) * 2**Fractional_shift_number * 2**Biais_shift_number)
Output_Layer_Biais[0][1] = int((NN.intercepts_[0][1]) * 2**Fractional_shift_number * 2**Biais_shift_number)
Output_Layer_Biais[1][0] = int((NN.intercepts_[1][0]) * 2**Fractional_shift_number * 2**Biais_shift_number)

print('Poids synaptiques = \n',Output_Layer_Coefs,'\n')
print('Biais = \n', Output_Layer_Biais,'\n')
```
